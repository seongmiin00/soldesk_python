```python
# 1/4 분기 매출액 합을 출력하는 프로그램 제작, int(), input()

# 1월 매출액:1000
# 2월 매출액:1000
# 3월 매출액:1000
# 1/4분기 매출액: 3000 만원
```

```python
jan = int(input('1월 매출액 : '))
feb = int(input('2월 매출액 : '))
mar = int(input('3월 매출액 : '))

qua_tot = jan + feb + mar
print(f'1/4분기 매출액 : {qua_tot} 만원')
```

```python
# m²/ 3.3 = 평

# 원룸의 width를 입력하세요:6
# 원룸의 height를 입력하세요:4
# 원룸의 크기: 24 m² 7.272727272727273 평
```

```python
width = int(input('원룸의 width를 입력하세요:'))
height = int(input('원룸의 height를 입력하세요:'))

pyeong = width * height / 3.3
print(f'원룸의 크기 : {width * height} m²{pyeong} 평')
```

```python
# 신체질량지수(BMI) = (몸무게(kg) / 신장의 제곱) * 10000

# 몸무게(kg):64
# 키(신장, cm):167
# BMI: 22.948115744558788

kg = int(input("몸무게(kg) : "))
cm = int(input("키(신장, cm) : "))

bmi = (kg / cm ** 2) * 10000
bmi = round(bmi,2)
print(f'BMI : {bmi}')
```

```python
# round 함수: 반올림
price = 10555.56755
print(round(price, 4))  # 5 에서는 반올림이 안됨
price = 10555.567550001
print(round(price, 4))  # 5 초과시 반올림 발생
```

```python
price = 10555.567550001
print(round(price, 3))
print(round(price, 2))
print(round(price, 1))
print(round(price, 0))
print(round(price, -1))
print(round(price, -2))
```

```python
print(round(1055, -1)) # 1060, 5에서 반올림
print(round(1055, -2)) # 1100
print(round(1050, -2)) # 1000
print(round(1051, -2)) # 1100, 50 초과시 반올림
print(round(10500, -3)) # 10000
print(round(10500.00001, -3)) # 11000, 500 초과시 반올림
```
