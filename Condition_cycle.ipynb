```python
# 1 ~ 1000까지의 수 중에서 2 and 3 and 4 and 5의 배수와 갯수를 출력하시오.
# hint: for, if

# 60  120  180  240  300  360  420  480  540  600  660  720  780  840  900  960
# cnt: 16

cnt = 0

for i in range(1, 1001):
    if i % 2 == 0 and i % 3 == 0 and i % 4 == 0 and i % 5 == 0:
        print(i, end=' ')
        cnt += 1
print('')
print(f'cnt : {cnt}')
```

```python
# 특정 단어를 입력 받아 순환 횟수 만큼 출력 할 것.
# 입력: 단어, 순환 횟수
# 최대 순환 횟수는 10회임, break를 이용하여 반복문 탈출(종료).

# 단어: ★
# 순환 횟수: 3
# ★★★

# 단어: ★
# 순환 횟수: 100
# ★★★★★★★★★★

word = '★'
cycle = int(input("순환 횟수 : "))

for i in range(cycle):
  print(word,end='')
  if i == 9:
    break
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# - for문 종료(탈출) 선언 필요, break
# - for문 순환 횟수는 100이상 자유롭게 지정: input 사용 필요 없음.

# 입력: 3
# 홀수
# 입력: 2
# 짝수
# 입력: 9999

for i in range(100):
  num = int(input("입력 : "))
  if num == 9999:
    print('종료')
    break
  if num % 2 == 0:
    result = '짝수'
  elif num % 2 == 1:
    result = '홀수'

  print(result)
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지 여부와 갯수를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# - 출력

# 입력: 1
# 홀수입니다.
# 짝수 갯수: 0 홀수 갯수: 1
# ------------------------------
# 입력: 2
# 짝수입니다.
# 짝수 갯수: 1 홀수 갯수: 1
# ------------------------------
# 입력: 3
# 홀수입니다.
# 짝수 갯수: 1 홀수 갯수: 2
# ------------------------------
# 입력: 9999
# 사용해 주셔서 감사합니다.

cnt1 = 0
cnt2 = 0

for i in range(100):
  num = int(input("입력 : "))
  if num == 9999:
    print('사용해 주셔서 감사합니다.')
    break
  if num % 2 == 0:
    result = '짝수입니다'
    cnt2 += 1
  elif num % 2 == 1:
    result = '홀수입니다'
    cnt1 += 1
  print(result)
  print(f'짝수 개수 : {cnt2} 홀수 개수 : {cnt1}')
```

```python
# 3번 순환하면서 숫자를 입력 받아, 입력 받은 숫자까지의 합을 출력 할 것.
# - 입력: 정수 1개
# - 출력:
#  입력: 1
#  합계: 1
#  입력: 3
#  합계: 4
#  입력: 5
#  합계: 9
sum = 0
for i in range(1,4):
  num = int(input("입력 : "))
  sum += num
  print(f'합계 : {sum}')
```

### 난수(random)
- 게임 분야
- 웹 개발: 이벤트, 사전에 지정이되면 안되는 다양성이 요구되는 기능
- 데이터분석: 바이러스 생성, 가상의 데이터 생성
- 각각의 언어는 난수 발생 알고리즘이 있어서 seed를 지정하면 같은
난수를 생성함.
- seed를 사용하는 경우는 가상의 데이터를 동일하게 생성하고, 그 데이터를 사용하는 로직의 성능을 개선시키는 경우 이용

```python
import random

random.seed(0)#동일 난수 생성 위치
for i in range(5):
  print(random.randint(1,5), end = " ")
```

```python
# 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것

# Hint: 난수, for, input, if
import random

com = random.randint(1,10)
print(com)

for i in range(3):
  user = int(input('예상되는 수 입력 : '))
  if user == com:
    print('맞습니다.')
    break
  elif user != com:
    print('틀렸습니다.')
print('최대 기회 3회를 모두 사용')
```

```python
# 힌트 기능 추가된 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것
# 숫자가 정답보다 크다면 [좀더 작은 수를 입력하세요.] 출력
# 숫자가 정답보다 작으면 [좀더 큰 수를 입력하세요.] 출력

# Hint: 난수, for, input, if

com = random.randint(1,10)
print(com)

for i in range(3):
  user = int(input('예상되는 수 입력 : '))
  if user == com:
    print('맞습니다.')
    break
  elif user != com:
    print('틀렸습니다.')
    if user > com:
      print('좀더 작은 수를 입력하세요.')
    elif user < com:
      print('좀더 큰 수를 입력하세요.')
    if i == 2:
      print('최대 기회 3회를 모두 사용')
```

```python
# 힌트 기능 추가된 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것
# 숫자가 정답보다 크다면 [좀더 작은 수를 입력하세요.] 출력
# 숫자가 정답보다 작으면 [좀더 큰 수를 입력하세요.] 출력
# 1회에 맞춘 경우 상금: 10만원
# 2회에 맞춘 경우 상금: 5만원
# 3회에 맞춘 경우 상금: 1만원

# 상금: 10만원
# 상금: 5만원
# 상금: 1만원
# 다음 기회에 시도해주세요.

# Hint: 난수, for, input, if

com = random.randint(1,10)
print(com)

for i in range(3):
  user = int(input('예상되는 수 입력 : '))
  if user == com:
    print('맞습니다.')
    if i == 0:
      price = '상금: 10만원'
    elif i == 1:
      price = '상금: 5만원'
    elif i == 2:
      price = '상금: 1만원'
    break
  elif user != com:
    print('틀렸습니다.')
    if user > com:
      print('좀더 작은 수를 입력하세요.')
    elif user < com:
      print('좀더 큰 수를 입력하세요.')
    if i == 2:
      print('최대 기회 3회를 모두 사용')
      price = '다음기회에 시도해주세요'

print(price)
```

### while 문
- 순환시 종료 시점을 알 수 없는 경우 사용
- 무한 반복이 필요한 경우, for는 반복 횟수를 알 수 있을 경우
- 네트워크 서버에서 주로 사용

```python
# 1 ~ 100 범위를 갖는 난수중에 50이 나오면 종료
while True:
  num = random.randint(1,100)
  if num == 50:
    break
  print(num, end=' ')
print(num)
```

```python
# 1 ~ 100 범위를 갖는 난수중에 50이 나오면 종료 난수는 한줄에 10개씩 출력
import time,random

cnt = 0
while True:
  cnt += 1
  time.sleep(0.5)
  num = random.randint(1,100)
  if num == 50:
    break
  if cnt < 10:
    print(num, end=' ')
  else:
    print(num)
    cnt = 0
```
